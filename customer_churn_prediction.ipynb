LABEL ENCODING


```python
for column in df:
    print(f'{column}:{df1[column].unique()}')
```

```python
#so basically we need to encode the object types to numeric for further prediction
# print all the objects
def print_unique_col_values(df):
    for column in df:
        if df[column].dtypes == 'object':
            print(f'{column}:{df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
# probabaly there are like no internet service, no phone service does comes under NO so we can normalize them
#just repplace with No
df1.replace('No internet service','No', inplace=True)
df1.replace('No phone service','No', inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
#replacing yes and no with 1 and 0

# Define a mapping dictionary
yes_no_to_01 = {'Yes': 1, 'No': 0}

# List of columns you want to convert
columns_to_convert = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']

# Loop through the columns and apply the mapping
for col in columns_to_convert:
    df1[col] = df1[col].map(yes_no_to_01)
```

```python
df1.info()
```

```python
for column in df:
    print(f'{column}:{df1[column].unique()}')
```

```python
#gender column
df1['gender'].replace({'Female':1,'Male':0}, inplace=True)
df1['gender'].unique()
```

ONE HOT ENCODING 

```python
#InternetService:['DSL' 'Fiber optic' 'No']  we have three unique values here for this we one hot encoding it will createthree different columns and makes it 0 and 1
#pd.get_dummies(data=df1,columns=['InternetService'])
```

```python
#we can do this remaining columns like contract and paymentmethod
df2=pd.get_dummies(data=df1, columns=['InternetService', 'Contract', 'PaymentMethod'])

df2.columns
```

FEATURE SCALING 


```python
#scaling tenure, MonthlyCharges, TotalCharges to make them 0 and 1 
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

TRAIN AND TEST SPLIT

```python
X = df2.drop('Churn', axis='columns')
y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=5)
```

NEURAL NETWORKS

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
    
])

model.compile(optimizer='adam',
             loss = 'binary_crossentropy',
             metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred[:10]
```

```python
y_test[:10]
```

```python
y_pred1 = []
for element in y_pred:
    if element > 0.5:
        y_pred1.append(1)
    else:
        y_pred1.append(0)
```

```python
y_pred1[:10]
```

```python
y_pred1
```

```python
#classification report
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred1))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels = y_test, predictions=y_pred1)
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot= True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#final prediction

final_y_pred = model.predict(X)
final_y_pred[:10]
```

```python
final_y_pred1 = []
for element in final_y_pred:
    if element > 0.5:
        final_y_pred1.append(1)
    else:
        final_y_pred1.append(0)
```
